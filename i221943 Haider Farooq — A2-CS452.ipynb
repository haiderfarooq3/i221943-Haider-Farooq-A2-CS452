```python
# ============================================================
# Legal Clause Similarity: Colab Notebook (revised, runs as-is)
# Baselines: (1) BiLSTM Siamese, (2) CNN + Self-Attention Siamese
# No pre-trained transformers; embeddings trained from scratch
# ============================================================

# --- Setup & Imports ---
!pip -q install kaggle==1.6.17

import os
import sys
import json
import math
import random
import zipfile
import glob
import gc
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support,
                             roc_auc_score, average_precision_score,
                             classification_report)

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, callbacks
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

print("Python", sys.version)
print("TensorFlow", tf.__version__)

# --- Reproducibility ---
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ============================================================
# 0) Data Access Helpers (Kaggle or Manual Upload)
# ============================================================

DATASET_SLUG = "bahushruth/legalclausedataset"
DATA_DIR = Path("data/legalclausedataset")
DATA_DIR.mkdir(parents=True, exist_ok=True)

def have_kaggle_token():
    """Return True if kaggle.json is available at ~/.kaggle/kaggle.json."""
    kaggle_json = Path.home() / ".kaggle" / "kaggle.json"
    return kaggle_json.exists()

def ensure_kaggle_token():
    """
    Ensure Kaggle API token exists; if not, prompt user to upload it in Colab.
    """
    kaggle_dir = Path.home() / ".kaggle"
    kaggle_dir.mkdir(exist_ok=True)
    kaggle_json = kaggle_dir / "kaggle.json"
    if not kaggle_json.exists():
        print("⏳ kaggle.json not found. Please upload your Kaggle API token.")
        from google.colab import files
        uploaded = files.upload()
        if "kaggle.json" not in uploaded:
            raise RuntimeError("kaggle.json not uploaded. Aborting.")
        with open(kaggle_json, "wb") as f:
            f.write(uploaded["kaggle.json"])
    os.chmod(kaggle_json, 0o600)
    print("✅ Kaggle token is set.")

def download_kaggle_dataset():
    """
    Download and unzip the Kaggle dataset into DATA_DIR using Kaggle CLI.
    """
    print("⬇️ Downloading dataset from Kaggle...")
    !kaggle datasets download -d $DATASET_SLUG -p $DATA_DIR --force
    zips = list(DATA_DIR.glob("*.zip"))
    if not zips:
        raise RuntimeError("Failed to download dataset zip from Kaggle.")
    for z in zips:
        with zipfile.ZipFile(z, 'r') as zip_ref:
            zip_ref.extractall(DATA_DIR)
        z.unlink()
    print("✅ Dataset downloaded and extracted to:", DATA_DIR)

def maybe_get_dataset():
    """
    Try Kaggle first. If not possible, allow user to manually upload CSVs.
    """
    csvs = list(DATA_DIR.glob("*.csv"))
    if csvs:
        print("✅ Found existing CSVs in", DATA_DIR)
        return

    # Try Kaggle path
    try:
        if not have_kaggle_token():
            ensure_kaggle_token()
        download_kaggle_dataset()
    except Exception as e:
        print(f"⚠️ Kaggle path failed: {e}")
        print("➡️ Option B: Please upload the CSV files manually.")
        from google.colab import files
        uploaded = files.upload()  # user selects multiple .csv files
        for name, content in uploaded.items():
            p = DATA_DIR / name
            with open(p, "wb") as f:
                f.write(content)
        print("✅ CSVs uploaded to:", DATA_DIR)

maybe_get_dataset()

# List CSVs we will use
csv_files = sorted([str(p) for p in DATA_DIR.glob("*.csv")])
print(f"Found {len(csv_files)} CSV files.")
print("\n".join(csv_files[:10]))

# ============================================================
# 1) Load & Clean Data
# ============================================================

def load_all_csvs(csv_paths):
    """
    Load all csvs from the dataset and unify columns to ['text', 'type', 'file'].
    """
    frames = []
    for path in csv_paths:
        df = pd.read_csv(path)
        # Normalize column names
        cols = [c.lower().strip() for c in df.columns]
        df.columns = cols

        # Try to detect text & type columns
        text_col = None
        for cand in ["clause", "text", "clause_text", "content"]:
            if cand in df.columns:
                text_col = cand
                break
        type_col = None
        for cand in ["type", "clause_type", "label", "category"]:
            if cand in df.columns:
                type_col = cand
                break

        if text_col is None or type_col is None:
            print(f"⚠️ Skipping {path} (couldn't find text/type columns)")
            continue

        tmp = df[[text_col, type_col]].copy()
        tmp.columns = ["text", "type"]
        tmp["file"] = Path(path).name.replace(".csv", "")
        frames.append(tmp)

    if not frames:
        raise RuntimeError("No valid CSVs found with recognizable columns.")

    all_df = pd.concat(frames, ignore_index=True)
    # Drop very short/empty texts
    all_df["text"] = all_df["text"].astype(str).str.strip()
    all_df = all_df[all_df["text"].str.len() > 5].reset_index(drop=True)
    return all_df

raw_df = load_all_csvs(csv_files)
print(raw_df.head())
print(raw_df["type"].value_counts().head())

# ------------------------------------------------------------
# Basic text cleaning. Keep it conservative for legal text.
# ------------------------------------------------------------
import re
def clean_text(s: str) -> str:
    s = s.lower()
    s = s.replace("\n", " ").replace("\t", " ")
    # normalize punctuation spacing but keep digits and %,$,§ etc.
    s = re.sub(r"[^a-z0-9\s\-\.,;:%/\(\)§$€£&']", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

raw_df["clean_text"] = raw_df["text"].map(clean_text)

# ============================================================
# 2) Create Pairwise Similarity Dataset
#    Positive = same 'type'; Negative = different 'type'
#    Balanced sampling; avoid explosion via capping per type
# ============================================================

def create_pairs(df: pd.DataFrame,
                 max_pos_pairs_per_type: int = 2000,
                 neg_ratio: float = 1.0,
                 seed: int = 42):
    """
    Create (text1, text2, label) pairs.
    - For each type, sample up to max_pos_pairs_per_type positive pairs.
    - For negatives, sample to match neg_ratio * num_positive (balanced by default).
    """
    rng = np.random.default_rng(seed)

    # Group indices by type
    type2idx = {}
    for t, sub in df.groupby("type"):
        type2idx[t] = sub.index.to_list()

    pairs_text1, pairs_text2, pairs_label = [], [], []

    # Positive pairs
    total_pos = 0
    for t, idxs in type2idx.items():
        if len(idxs) < 2:
            continue
        # For each index, pick one partner at random to build ~linear pairs
        pos_candidates = []
        for i in idxs:
            j = int(rng.choice([k for k in idxs if k != i]))
            pos_candidates.append((i, j))
        rng.shuffle(pos_candidates)
        pos_candidates = pos_candidates[:max_pos_pairs_per_type]

        for i, j in pos_candidates:
            pairs_text1.append(df.loc[i, "clean_text"])
            pairs_text2.append(df.loc[j, "clean_text"])
            pairs_label.append(1)
        total_pos += len(pos_candidates)

    # Negative pairs
    types = list(type2idx.keys())
    neg_needed = int(total_pos * neg_ratio)
    for _ in range(neg_needed):
        t1, t2 = rng.choice(types, size=2, replace=False)
        i = int(rng.choice(type2idx[t1]))
        j = int(rng.choice(type2idx[t2]))
        pairs_text1.append(df.loc[i, "clean_text"])
        pairs_text2.append(df.loc[j, "clean_text"])
        pairs_label.append(0)

    pair_df = pd.DataFrame({
        "text1": pairs_text1,
        "text2": pairs_text2,
        "label": pairs_label
    })
    return pair_df

# --- Split clauses first (to avoid leakage across splits) ---
train_clauses, test_clauses = train_test_split(
    raw_df.index, test_size=0.15, random_state=SEED, shuffle=True
)
train_clauses, val_clauses = train_test_split(
    train_clauses, test_size=0.1765, random_state=SEED
)
# Train ~70%, Val ~15%, Test ~15%

train_df = raw_df.loc[train_clauses].reset_index(drop=True)
val_df   = raw_df.loc[val_clauses].reset_index(drop=True)
test_df  = raw_df.loc[test_clauses].reset_index(drop=True)

print(f"Clauses -> Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")

# Build pairs within each split
train_pairs = create_pairs(train_df, max_pos_pairs_per_type=1500, neg_ratio=1.0, seed=SEED)
val_pairs   = create_pairs(val_df,   max_pos_pairs_per_type=400,  neg_ratio=1.0, seed=SEED+1)
test_pairs  = create_pairs(test_df,  max_pos_pairs_per_type=400,  neg_ratio=1.0, seed=SEED+2)

print("Pairs -> Train:", train_pairs.shape, "Val:", val_pairs.shape, "Test:", test_pairs.shape)
print(train_pairs["label"].value_counts())

# ============================================================
# 3) Tokenization & Vectorization
# ============================================================

# Fit tokenizer on TRAIN ONLY (best practice)
tokenizer = Tokenizer(num_words=30000, oov_token="<OOV>")
tokenizer.fit_on_texts(train_df["clean_text"].tolist())

def vectorize_pairs(pair_df: pd.DataFrame, tokenizer, max_len: int = None):
    s1 = tokenizer.texts_to_sequences(pair_df["text1"].tolist())
    s2 = tokenizer.texts_to_sequences(pair_df["text2"].tolist())
    if max_len is None:
        # heuristic: 95th percentile length on train
        lengths = [len(x) for x in s1 + s2]
        max_len = int(np.percentile(lengths, 95))
        max_len = max(32, min(512, max_len))
    x1 = pad_sequences(s1, maxlen=max_len, padding='post', truncating='post')
    x2 = pad_sequences(s2, maxlen=max_len, padding='post', truncating='post')
    y  = pair_df["label"].astype(int).values
    return (x1, x2, y, max_len)

X1_train, X2_train, y_train, MAX_LEN = vectorize_pairs(train_pairs, tokenizer, max_len=None)
X1_val,   X2_val,   y_val,   _       = vectorize_pairs(val_pairs, tokenizer, max_len=MAX_LEN)
X1_test,  X2_test,  y_test,  _       = vectorize_pairs(test_pairs, tokenizer, max_len=MAX_LEN)

VOCAB_SIZE = min(30000, len(tokenizer.word_index) + 1)
print(f"Vocab size: {VOCAB_SIZE}, Max length: {MAX_LEN}")

# ============================================================
# 4) Model Building Blocks (Encoders + Siamese Head)
# ============================================================

class SiameseCombine(layers.Layer):
    """
    Combine two encoded vectors u, v using |u-v|, u*v, u, v (popular for similarity).
    """
    def call(self, inputs):
        u, v = inputs
        diff = tf.math.abs(u - v)
        prod = u * v
        return tf.concat([diff, prod, u, v], axis=-1)

def build_siamese_classifier(encoder: tf.keras.Model,
                             embed_dim: int,
                             l2: float = 1e-5,
                             dropout: float = 0.2):
    """
    Wrap a sentence encoder into a Siamese architecture for pair classification.
    """
    inp1 = layers.Input(shape=(MAX_LEN,), name="s1")
    inp2 = layers.Input(shape=(MAX_LEN,), name="s2")

    u = encoder(inp1)
    v = encoder(inp2)

    x = SiameseCombine()([u, v])
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(256, activation="relu",
                     kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(64, activation="relu",
                     kernel_regularizer=regularizers.l2(l2))(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    model = models.Model(inputs=[inp1, inp2], outputs=out)
    return model

# ------------------------------------------------------------
# Encoder A: BiLSTM (baseline)
# ------------------------------------------------------------
def build_encoder_bilstm(vocab_size: int, embed_dim: int = 128, lstm_units: int = 128):
    inp = layers.Input(shape=(MAX_LEN,), name="tokens")
    x = layers.Embedding(vocab_size, embed_dim, mask_zero=True, name="emb")(inp)
    x = layers.Bidirectional(layers.LSTM(lstm_units, return_sequences=True))(x)
    x = layers.GlobalMaxPooling1D()(x)
    x = layers.Dense(128, activation="relu")(x)
    enc = models.Model(inp, x, name="BiLSTM_Encoder")
    return enc

# ------------------------------------------------------------
# Simple Additive Attention (Keras-3 safe)
# ------------------------------------------------------------
class AdditiveAttention(layers.Layer):
    def __init__(self, hidden_dim=128):
        super().__init__()
        self.W = layers.Dense(hidden_dim, activation="tanh")
        self.v = layers.Dense(1, activation=None)

    def call(self, h, mask=None):
        # h: (B, T, D)
        score = self.v(self.W(h))  # (B, T, 1)
        if mask is not None:
            mask = tf.cast(mask, tf.float32)  # (B, T)
            mask = tf.expand_dims(mask, -1)   # (B, T, 1)
            score = score - 1e9 * (1.0 - mask)
        attn = tf.nn.softmax(score, axis=1)  # (B, T, 1)
        context = tf.reduce_sum(attn * h, axis=1)  # (B, D)
        return context

# ------------------------------------------------------------
# Encoder B: CNN + Self-Attention (non-transformer, Keras-3 safe)
# ------------------------------------------------------------
def build_encoder_cnn_attention(vocab_size: int,
                                embed_dim: int = 128,
                                num_filters: int = 128):
    inp = layers.Input(shape=(MAX_LEN,), name="tokens")
    emb = layers.Embedding(vocab_size, embed_dim, mask_zero=True, name="emb")(inp)
    # Conv blocks (Kim CNN style). Convs ignore masks; provide our own mask to attention.
    conv_3 = layers.Conv1D(num_filters, 3, padding="same", activation="relu")(emb)
    conv_4 = layers.Conv1D(num_filters, 4, padding="same", activation="relu")(emb)
    conv_5 = layers.Conv1D(num_filters, 5, padding="same", activation="relu")(emb)
    x = layers.Concatenate()([conv_3, conv_4, conv_5])  # (B, T, num_filters*3)

    # Keras-3 SAFE: compute padding mask inside a Keras layer
    pad_mask = layers.Lambda(lambda t: tf.cast(tf.not_equal(t, 0), tf.float32),
                             name="pad_mask")(inp)      # (B, T)

    context = AdditiveAttention(hidden_dim=128)(x, mask=pad_mask)
    x = layers.Dense(128, activation="relu")(context)
    enc = models.Model(inp, x, name="CNN_Attn_Encoder")
    return enc

# ============================================================
# 5) Train & Evaluate utilities
# ============================================================

def compile_model(model: tf.keras.Model, lr=2e-3):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss="binary_crossentropy",
        metrics=[
            tf.keras.metrics.BinaryAccuracy(name="acc"),
            tf.keras.metrics.AUC(name="auc", curve="ROC"),
            tf.keras.metrics.AUC(name="auc_pr", curve="PR"),
            tf.keras.metrics.Precision(name="prec"),
            tf.keras.metrics.Recall(name="recall"),
        ],
    )
    return model

def train_model(model_builder,
                encoder_name: str,
                epochs=8,
                batch_size=128,
                lr=2e-3):
    print(f"\n==================== Training: {encoder_name} ====================")
    encoder = model_builder(VOCAB_SIZE)
    model = build_siamese_classifier(encoder, embed_dim=128)
    model = compile_model(model, lr=lr)

    es = callbacks.EarlyStopping(monitor="val_auc", patience=2, mode="max", restore_best_weights=True)
    rlrop = callbacks.ReduceLROnPlateau(monitor="val_auc", factor=0.5, patience=1, mode="max", min_lr=1e-5)

    hist = model.fit(
        [X1_train, X2_train], y_train,
        validation_data=([X1_val, X2_val], y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[es, rlrop],
        verbose=1
    )
    return model, hist

def evaluate_and_report(model, X1, X2, y_true, title="Model"):
    probs = model.predict([X1, X2], batch_size=256, verbose=0).ravel()
    preds = (probs >= 0.5).astype(int)

    acc = accuracy_score(y_true, preds)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, preds, average="binary", zero_division=0)
    try:
        roc = roc_auc_score(y_true, probs)
    except:
        roc = float("nan")
    try:
        pr_auc = average_precision_score(y_true, probs)
    except:
        pr_auc = float("nan")

    print(f"\n=== {title} ===")
    print(f"Accuracy  : {acc:.4f}")
    print(f"Precision : {prec:.4f}")
    print(f"Recall    : {rec:.4f}")
    print(f"F1-Score  : {f1:.4f}")
    print(f"ROC-AUC   : {roc:.4f}")
    print(f"PR-AUC    : {pr_auc:.4f}")
    print("\nClassification report:")
    print(classification_report(y_true, preds, digits=4))

    return {
        "accuracy": acc, "precision": prec, "recall": rec, "f1": f1,
        "roc_auc": roc, "pr_auc": pr_auc, "probs": probs, "preds": preds
    }

def plot_roc_pr(y_true, probs, label="Model"):
    from sklearn.metrics import roc_curve, precision_recall_curve
    fpr, tpr, _ = roc_curve(y_true, probs)
    prec, rec, _ = precision_recall_curve(y_true, probs)

    plt.figure(figsize=(5,4))
    plt.plot(fpr, tpr, label=f"{label} (AUC={roc_auc_score(y_true, probs):.3f})")
    plt.plot([0,1], [0,1], linestyle="--")
    plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title("ROC Curve")
    plt.legend(); plt.grid(True); plt.show()

    plt.figure(figsize=(5,4))
    plt.plot(rec, prec, label=f"{label} (AP={average_precision_score(y_true, probs):.3f})")
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("Precision-Recall Curve")
    plt.legend(); plt.grid(True); plt.show()

def show_qualitative_examples(df_pairs, probs, preds, k=5):
    """
    Show a few correct/incorrect matches for qualitative analysis.
    """
    df = df_pairs.copy()
    df["prob"] = probs
    df["pred"] = preds
    df["correct"] = (df["pred"] == df["label"]).astype(int)

    print("\n✅ Correct Positive (True Similar) examples:")
    ex = df[(df["label"]==1) & (df["pred"]==1)].sort_values("prob", ascending=False).head(k)
    for _, row in ex.iterrows():
        print("-"*80)
        print(f"Score: {row['prob']:.3f}")
        print("Clause A:", row["text1"])
        print("Clause B:", row["text2"])

    print("\n✅ Correct Negative (True Different) examples:")
    ex = df[(df["label"]==0) & (df["pred"]==0)].sort_values("prob", ascending=True).head(k)
    for _, row in ex.iterrows():
        print("-"*80)
        print(f"Score: {row['prob']:.3f}")
        print("Clause A:", row["text1"])
        print("Clause B:", row["text2"])

    print("\n❌ False Positives (Predicted Similar but Different):")
    ex = df[(df["label"]==0) & (df["pred"]==1)].sort_values("prob", ascending=False).head(k)
    for _, row in ex.iterrows():
        print("-"*80)
        print(f"Score: {row['prob']:.3f}")
        print("Clause A:", row["text1"])
        print("Clause B:", row["text2"])

    print("\n❌ False Negatives (Predicted Different but Similar):")
    ex = df[(df["label"]==1) & (df["pred"]==0)].sort_values("prob", ascending=True).head(k)
    for _, row in ex.iterrows():
        print("-"*80)
        print(f"Score: {row['prob']:.3f}")
        print("Clause A:", row["text1"])
        print("Clause B:", row["text2"])

# ============================================================
# 6) Train the Two Baselines
# ============================================================

# Baseline 1: BiLSTM
bilstm_model, hist1 = train_model(build_encoder_bilstm, "BiLSTM", epochs=8, batch_size=128, lr=2e-3)
res1 = evaluate_and_report(bilstm_model, X1_test, X2_test, y_test, title="BiLSTM (Test)")
plot_roc_pr(y_test, res1["probs"], label="BiLSTM")

# Baseline 2: CNN + Self-Attention (Keras-3 safe)
cnn_attn_model, hist2 = train_model(build_encoder_cnn_attention, "CNN+SelfAttention", epochs=8, batch_size=128, lr=2e-3)
res2 = evaluate_and_report(cnn_attn_model, X1_test, X2_test, y_test, title="CNN+SelfAttention (Test)")
plot_roc_pr(y_test, res2["probs"], label="CNN+SelfAttn")

# ============================================================
# 7) Quantitative Comparison
# ============================================================

def compare_results(r1, r2, name1="BiLSTM", name2="CNN+SelfAttn"):
    table = pd.DataFrame([
        {"Model": name1, "Accuracy": r1["accuracy"], "Precision": r1["precision"], "Recall": r1["recall"],
         "F1": r1["f1"], "ROC-AUC": r1["roc_auc"], "PR-AUC": r1["pr_auc"]},
        {"Model": name2, "Accuracy": r2["accuracy"], "Precision": r2["precision"], "Recall": r2["recall"],
         "F1": r2["f1"], "ROC-AUC": r2["roc_auc"], "PR-AUC": r2["pr_auc"]},
    ])
    display(table.style.format("{:.4f}"))
    return table

_ = compare_results(res1, res2)

# ============================================================
# 8) Qualitative Examples
# ============================================================

print("\n==== Qualitative Examples for BiLSTM ====")
show_qualitative_examples(test_pairs, res1["probs"], res1["preds"], k=5)

print("\n==== Qualitative Examples for CNN+SelfAttention ====")
show_qualitative_examples(test_pairs, res2["probs"], res2["preds"], k=5)

# ============================================================
# 9) Brief Discussion (printed)
# ============================================================

def discussion(r1, r2):
    better_roc = "BiLSTM" if r1["roc_auc"] >= r2["roc_auc"] else "CNN+SelfAttention"
    better_pr  = "BiLSTM" if r1["pr_auc"] >= r2["pr_auc"] else "CNN+SelfAttention"
    better_f1  = "BiLSTM" if r1["f1"] >= r2["f1"] else "CNN+SelfAttention"

    print("\n=== Comparative Analysis (Auto-generated notes) ===")
    print(f"- Overall ROC-AUC winner : {better_roc}")
    print(f"- Overall PR-AUC winner  : {better_pr}")
    print(f"- Overall F1 winner      : {better_f1}")
    print("\nStrengths & Weaknesses (typical trends):")
    print("* BiLSTM:")
    print("  - Strengths: captures sequential dependencies and long-ish context; robust on rephrasing.")
    print("  - Weaknesses: may overfit on surface patterns with limited data; slower than CNN at inference.")
    print("* CNN+Self-Attention:")
    print("  - Strengths: strong n-gram feature extraction, attention highlights salient tokens; faster and lighter.")
    print("  - Weaknesses: may miss long-range order compared to BiLSTM; attention here is shallow (non-transformer).")
    print("\nRecommendation:")
    print("- If latency matters or you want simpler deployment, CNN+Self-Attention is attractive.")
    print("- If you need richer sequential modeling, BiLSTM may edge out on nuanced paraphrases.")
    print("- Ensembling both (e.g., probability averaging) often improves robustness.")

discussion(res1, res2)

print("\n✅ Done.")
```
